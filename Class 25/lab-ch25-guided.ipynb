### Q4: Prompt engineering lessons

*What makes a good prompt for Claude Code? What makes a bad one? Give one specific example of a prompt you improved.*

**Your answer:**
I think a good prompt for claude is one that is specific and doesn't leave room for claude to think on its own. The more general a prompt, the more claude thinks on its own and instead of doing what you ask, it'll try and add on and improve which sometimes works but sometimes doesn't. A prompt I used was "make the overall theme a minimalistic theme", but I changed it on and added "make the theme and black clean minimalist theme", while I gave it room to move with, I also specified the colors I wanted which I felt was the most important part, and gave it no room to change on in regards for that.
